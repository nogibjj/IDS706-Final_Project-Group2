```python
df_train_copy = df_train.copy()


#drop unnecessary columns
df_train_copy[['Date','Time']] = df_train_copy.trans_date_trans_time.str.split(" ",expand=True)
df_train_copy.head()


#switch the columns in the dataframe
df_train_copy = df_train_copy[['category', 'amt', 'city_pop','city','state', 'merch_lat', 'merch_long', 'is_fraud', 'Date', 'Time']]


df_train_copy = df_train_copy[['Date', 'Time','category', 'amt', 'city_pop','city','state', 'is_fraud',]]

#aggregate by date and state
df_train_copy['Date'] = pd.to_datetime(df_train_copy['Date'])





##aggregate by date and state





#= df_train_copy.groupby(['Date','state']).agg({'amt':'sum','is_fraud':'sum'}).reset_index()


df_train_copy.shape


#df_train_copy = df_train_copy.drop(columns)
```

### Data cleaned only

```python
## Transfer to csv

df_train_copy
```

```python
# create a new column for fraud amount where is_fraud = 1 and put the amount in the column, else put 0
df_train_copy['fraud_amt'] = np.where(df_train_copy['is_fraud'] == 1, df_train_copy['amt'], 0)


df_train_copy
```

### 1. Data aggregated on daily level + count of frauds each day - df1

```python
#df_train_copy.head(20)


df1 = df_train_copy.copy()

#df_train_copy['Date'] = 2019-01-01'].value_counts()

#df1= df1.groupby(['Date'])['is_fraud'].sum().reset_index()


df1 = df1.groupby('Date').size().reset_index(name='count')




df1.rename(columns ={'count' : 'Daily transactions Count'}, inplace=True)


df1
```

### 1. Data aggregated on daily level + count of frauds each day - Daily (df1 and df2 taken)

```python
df2 = df_train_copy.copy()

#df2 = df2.groupby(['Date']).agg({'is_fraud':'sum'}).reset_index()

df2 = df2.groupby(['Date']).agg({'amt':'sum','is_fraud':'sum', 'fraud_amt':'sum'}).reset_index()

df2
```

```python

Daily = df2.merge(df1, on=["Date"], how="outer")



#Daily 


Daily.rename(columns ={'amt' : 'Daily transactions amount'}, inplace=True)

Daily.rename(columns ={'is_fraud' : 'Daily Fraudulent Transactions'}, inplace=True)

Daily.rename(columns ={'fraud_amt' : 'Daily Fraudulent amount'}, inplace=True)


Daily.head()
```

### Transform to CSV

```python
Daily.head()

Daily.to_csv("Daily_train_cleaned.csv", index= False)



#uber_merged.to_csv("uber_merged.csv", index= False)
```

### 2. Aggregated by date, category

```python

df_category = df_train_copy.copy()

df_category = df_category.groupby(['Date','category']).agg({'amt':'sum','is_fraud':'sum','fraud_amt': 'sum'}).reset_index()

df_category.rename(columns ={'amt' : 'transactions amount'}, inplace=True)

df_category.rename(columns ={'is_fraud' : 'Fraud count'}, inplace=True)

df_category.rename(columns ={'fraud_amt' : 'Fraud amount per category'}, inplace=True)

df_category
```

### Transform to CSV

```python
df_category.head()

df_category.to_csv("agg_daily_category_train_cleaned.csv", index= False)


```

### 3. Aggregated by date, state(training)

```python

##aggregate by date and state


df_state = df_train_copy.copy()

df_state = df_state.groupby(['Date','state']).agg({'amt':'sum','is_fraud':'sum','fraud_amt': 'sum'}).reset_index()


#df_train_copy = df_train_copy.groupby(['Date','state']).agg({'amt':'sum','is_fraud':'sum'}).reset_index()






#= df_train_copy.groupby(['Date','state']).agg({'amt':'sum','is_fraud':'sum'}).reset_index()
# 

df_state.rename(columns ={'amt' : 'transactions amount'}, inplace=True)

df_state.rename(columns ={'is_fraud' : 'Fraud count per state'}, inplace=True)

df_state.rename(columns ={'fraud_amt' : 'Fraud amount per state'}, inplace=True)

df_state

```

### Transform to CSV
